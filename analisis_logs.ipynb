```python
import matplotlib.pyplot as plt
from datetime import datetime

usuarios_error = {}
errores_por_hora = {}

with open("logs.txt", "r") as archivo:
    for linea in archivo:
        if linea.strip():  # Procesa solo si la línea no está vacía
            fecha_str, usuario, estado = linea.strip().split(", ")
            fecha = datetime.strptime(fecha_str, "%Y-%m-%d %H:%M:%S")
            hora = fecha.strftime("%H")

            if estado == "ERROR":
                usuarios_error[usuario] = usuarios_error.get(usuario, 0) + 1
                errores_por_hora[hora] = errores_por_hora.get(hora, 0) + 1

print("usuarios_error:", usuarios_error)
print("errores_por_hora:", errores_por_hora)
```

```python
plt.figure(figsize=(10, 4))
           
plt.subplot(1,2,1)
plt.bar(usuarios_error.keys(), usuarios_error.values(), color='red')
plt.xlabel('Usuario')
plt.ylabel('Número de errores')
plt.title('Errores de acceso por usuario')

plt.subplot(1,2,2)
plt.bar(errores_por_hora.keys(), errores_por_hora.values(), color='blue')
plt.xlabel('Hora')
plt.ylabel('Número de errores')
plt.title('Errores de acceso por hora')

plt.tight_layout()
plt.show()
```
